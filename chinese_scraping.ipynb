```python
import os
import pandas as pd
import json
import pathlib

dataFolder = './dataJson'
files = os.listdir(dataFolder)
files
```

```python
masterDF = pd.DataFrame()

for f in files:
    if f.endswith('.json'):
        with open(os.path.join(dataFolder, f)) as ff:
            data = ff.read() # this gives us str
            result = json.loads(data)
            # print(result.keys())
            print(result['organic_results'])
            print(len(result['organic_results']))
            df = pd.DataFrame(result['organic_results'])
            # display(df)
            masterDF = pd.concat([masterDF, df], ignore_index=True)

            print('-----\n')

display(masterDF)
```

```python
# create a long long list of word from the first 5 hk links

from urllib.request import urlopen
import urllib.request
from bs4 import BeautifulSoup as soup
import time
import jieba
import jieba.analyse as jaa

htmlTagToKeep = ['title', 'p', 'h1', 'h2', 'h3', 'h4']

hk_sentences = [] # a lot of sentence from html tags
tw_sentences = [] # a lot of sentence from html tags

for index, url in enumerate(hklinks['link']):
    if index < 30:
        try:
            print(index, url)

            hdr = {'User-Agent':'Mozilla/5.0', 'Accept': 'text/html, application/xhtml+xml, application/xml;q=0.9,*/*;q=0.8'}
            req = urllib.request.Request(url, headers=hdr)
            html_page = urllib.request.urlopen(req, timeout=10) # open an url
            html_soup = soup(html_page, "html.parser") # use bs to create a soup object for th html page

            all_tags = html_soup.findAll(htmlTagToKeep) # extract all the important tags
            for tag in all_tags:
                # print(tag.text)
                hk_sentences.append(tag.text)
        except Exception as err:
            print('error : ', err)

for index, url in enumerate(twlinks['link']):
    if index < 30:
        try:
            print(index, url)

            hdr = {'User-Agent':'Mozilla/5.0', 'Accept': 'text/html, application/xhtml+xml, application/xml;q=0.9,*/*;q=0.8'}
            req = urllib.request.Request(url, headers=hdr)
            html_page = urllib.request.urlopen(req, timeout=10) # open an url
            html_soup = soup(html_page, "html.parser") # use bs to create a soup object for th html page

            all_tags = html_soup.findAll(htmlTagToKeep) # extract all the important tags
            for tag in all_tags:
                # print(tag.text)
                tw_sentences.append(tag.text)
        except Exception as err:
            print('error : ', err)

# hk_sentences
```

```python
hk_keywords = [] # the long long list of keywords
tw_keywords = [] # the long long list of keywords

#  create the long long list of keywords by appending output from jaa
for index, sentence in enumerate(hk_sentences):
    # print(index, sentence)
    kaa_keywords_list = jaa.extract_tags(sentence, topK=10)
    for keyword in kaa_keywords_list:
        hk_keywords.append(keyword)
for index, sentence in enumerate(tw_sentences):
    # print(index, sentence)
    kaa_keywords_list = jaa.extract_tags(sentence, topK=10)
    for keyword in kaa_keywords_list:
        tw_keywords.append(keyword)

len(hk_keywords)
len(tw_keywords)

f'HK keywords: {len(hk_keywords)} vs TW keywords:{len(tw_keywords)}'

```

```python
hk_keywords = [] # the long long list of keywords
tw_keywords = [] # the long long list of keywords

#  create the long long list of keywords by appending output from jaa
for index, sentence in enumerate(hk_sentences):
    # print(index, sentence)
    kaa_keywords_list = jaa.extract_tags(sentence, topK=30)
    for keyword in kaa_keywords_list:
        hk_keywords.append(keyword)
for index, sentence in enumerate(tw_sentences):
    # print(index, sentence)
    kaa_keywords_list = jaa.extract_tags(sentence, topK=30)
    for keyword in kaa_keywords_list:
        tw_keywords.append(keyword)

len(hk_keywords)
len(tw_keywords)

f'HK keywords: {len(hk_keywords)} vs TW keywords:{len(tw_keywords)}'
```
